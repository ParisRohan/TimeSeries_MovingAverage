```python
#Load the required libraries
import pandas_datareader as pdr
import pandas as pd
```

```python
## Read Facebook stocks data from yahoo
df_fb = pdr.get_data_yahoo('FB')
df_fb.head()
```

```python
## Read Amazon stocks data from yahoo
df_amazon = pdr.get_data_yahoo('AMZN')
df_amazon.head()
```

```python
## Read Apple stocks data from yahoo
df_apple = pdr.get_data_yahoo('AAPL')
df_apple.head()
```

```python
## Read Netflix stocks data from yahoo
df_netflix = pdr.get_data_yahoo('NFLX')
df_netflix.head()
```

```python
## Read Google stocks data from yahoo
df_google = pdr.get_data_yahoo('GOOGL')
df_google.head()
```

```python
#Reset index as the current index is the Date feature
df_fb=df_fb.reset_index()
df_amazon=df_amazon.reset_index()
df_apple=df_apple.reset_index()
df_netflix=df_netflix.reset_index()
df_google=df_google.reset_index()
```

```python
#Get the covid-19 pandemic data from 01-Jan-2020 to 01-Apr-2022
df_fb=df_fb.iloc[683:1251]
df_amazon=df_amazon.iloc[683:1251]
df_apple=df_apple.iloc[683:1251]
df_netflix=df_netflix.iloc[683:1251]
df_google=df_google.iloc[683:1251]
```

```python
#Merge the different companies datasets
df_faang=pd.concat([df_fb.assign(Name="Facebook"), df_amazon.assign(Name="Amazon"), df_apple.assign(Name="Apple")
                    , df_netflix.assign(Name="Netflix"), df_google.assign(Name="Google")
                   ])
df_faang.head()
```

```python
# saving the dataframe
df_faang.to_csv('faang_stocks_pandemic_data.csv')
```
